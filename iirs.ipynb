#**Working with multiple plots**

The subplot() command specifies numrows,numcols,plot_number where plot_number ranges from 1 to numrows*numcols You can creare an arbitray number of subplots

```python
x=np.arange(0,10,0.2)
```

```python
y_sin=np.sin(x)
y_cos=np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine Plot')
```

```python
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cos plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y_sin)
plt.title('Sine plot')

plt.subplot(1,2,2)
plt.plot(x,y_cos)
plt.title('Cos Plot')
```

```python
plt.close()
```
